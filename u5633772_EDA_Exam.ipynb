```python
import matplotlib.pyplot as plt
import seaborn as sns

top_5 = avg_life_expectancy_by_country.sort_values(by='Life Expectancy', ascending=False).head(5)
worst_5 = avg_life_expectancy_by_country.sort_values(by='Life Expectancy', ascending=True).head(5)

# Add a category column to distinguish top and worst
top_5['Category'] = 'Top 5'
worst_5['Category'] = 'Worst 5'

# Concatenate the two dataframes
combined_df = pd.concat([top_5, worst_5])

plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Life Expectancy', hue='Category', data=combined_df, palette={'Top 5': 'green', 'Worst 5': 'red'})
plt.title('Top 5 and Worst 5 Countries by Average Life Expectancy')
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Define the columns related to diet calories and rice consumption
diet_consumption_columns = [
    'Diet Calories Animal Protein',
    'Diet Calories Plant Protein',
    'Diet Calories Fat',
    'Diet Calories Carbohydrates',
    'Cereal Consumption Rice'
]

# Calculate the average of these consumption columns for each country
avg_consumption_by_country = df.groupby('Country')[diet_consumption_columns].mean().reset_index()

# Get the list of top 5 countries from the previously generated 'top_5' DataFrame
top_5_countries_list = top_5['Country'].tolist()

# Filter the average consumption data for these top 5 countries
top_5_diet_data = avg_consumption_by_country[avg_consumption_by_country['Country'].isin(top_5_countries_list)].copy()

# Add a 'Continent' column based on the country
continent_map = {
    'San Marino': 'Europe',
    'Monaco': 'Europe',
    'Japan': 'Asia',
    'Hong Kong': 'Asia',
    'Macao': 'Asia'
}
top_5_diet_data['Continent'] = top_5_diet_data['Country'].map(continent_map)

# Melt the DataFrame to prepare it for plotting multiple consumption types
melted_diet_data = top_5_diet_data.melt(
    id_vars=['Country', 'Continent'],
    value_vars=diet_consumption_columns,
    var_name='Consumption Type',
    value_name='Average Consumption'
)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 8))
sns.barplot(
    x='Country',
    y='Average Consumption',
    hue='Consumption Type',
    data=melted_diet_data,
    palette='tab10',
    order=[
        'San Marino', 'Monaco', # Europe
        'Japan', 'Hong Kong', 'Macao' # Asia
    ]
)
plt.title('Average Diet Calories and Rice Consumption for Top 5 Life Expectancy Countries')
plt.xlabel('Country')
plt.ylabel('Average Consumption (units vary by type)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Consumption Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Compare the common diets of countries with high life expectancy by filtering for countries with complete diet data across 'Diet Calories Animal Protein', 'Diet Calories Plant Protein', 'Diet Calories Fat', 'Diet Calories Carbohydrates', and 'Cereal Consumption Rice' columns, then selecting the top 5 of these countries with the highest life expectancy, visualizing their dietary composition using a bar chart, and summarizing the findings.

## Identify Countries with Complete Diet Data

### Subtask:
Filter the `avg_consumption_by_country` DataFrame to include only countries that have non-null values for all the specified diet calorie and rice consumption columns. This will ensure we are working with complete data for dietary analysis.


**Reasoning**:
First, I will create a list of the diet consumption columns that need to be checked for non-null values. Then, I will filter the `avg_consumption_by_country` DataFrame to include only countries where all these specified columns have complete (non-null) data.



```python
diet_columns_for_filter = [
    'Diet Calories Animal Protein',
    'Diet Calories Plant Protein',
    'Diet Calories Fat',
    'Diet Calories Carbohydrates',
    'Cereal Consumption Rice'
]

filtered_diet_data = avg_consumption_by_country.dropna(subset=diet_columns_for_filter)

print("Shape of original avg_consumption_by_country:", avg_consumption_by_country.shape)
print("Shape of filtered_diet_data:", filtered_diet_data.shape)
display(filtered_diet_data.head())
```

**Reasoning**:
Now that we have filtered the diet data to include only countries with complete information, the next step is to identify the top 5 countries with the highest life expectancy from this filtered set. To do this, I will merge the `filtered_diet_data` with `avg_life_expectancy_by_country` and then sort by life expectancy to get the top 5.



```python
merged_data = pd.merge(filtered_diet_data, avg_life_expectancy_by_country, on='Country', how='inner')

top_5_complete_diet_countries = merged_data.sort_values(by='Life Expectancy', ascending=False).head(5)

print("Top 5 countries with complete diet data and highest life expectancy:")
display(top_5_complete_diet_countries)
```

**Reasoning**:
To visualize the dietary composition of the top 5 countries with complete diet data and highest life expectancy, I need to melt the `top_5_complete_diet_countries` DataFrame. This will transform the diet consumption columns into a single 'Consumption Type' column, making it suitable for plotting with a bar chart.



```python
diet_columns_for_plot = [
    'Diet Calories Animal Protein',
    'Diet Calories Plant Protein',
    'Diet Calories Fat',
    'Diet Calories Carbohydrates',
    'Cereal Consumption Rice'
]

melted_top_5_diet_data = top_5_complete_diet_countries.melt(
    id_vars=['Country', 'Life Expectancy'],
    value_vars=diet_columns_for_plot,
    var_name='Consumption Type',
    value_name='Average Consumption'
)

print("Melted data for top 5 countries (head):")
display(melted_top_5_diet_data.head())
```

**Reasoning**:
Now that the data is melted, I will create a bar chart to visualize the average diet calories and rice consumption for the top 5 countries with complete diet data and highest life expectancy, allowing for a clear comparison across different consumption types and countries.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))
sns.barplot(
    x='Country',
    y='Average Consumption',
    hue='Consumption Type',
    data=melted_top_5_diet_data,
    palette='viridis'
)
plt.title('Average Diet Calories and Rice Consumption for Top 5 Life Expectancy Countries (Complete Data)')
plt.xlabel('Country')
plt.ylabel('Average Consumption (kcal or kg/year)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Consumption Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Summary of Findings for Top 5 Countries with Complete Diet Data and Highest Life Expectancy

The bar chart visualizes the average consumption of Animal Protein, Plant Protein, Fat, Carbohydrates, and Rice for the top 5 countries (Japan, Switzerland, Iceland, Australia, Italy) that have complete dietary data and the highest life expectancy.
